```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('hr_data.csv')
```

1. Гістограма фірми з даними про вік працівників

```python
sns.histplot(data=df, x='Age')
```

2. Загальний графік розсіювання по продуктивності працівників, посади та зарплатні

```python
sns.scatterplot(x="Age",
    y="Productivity (%)",
    size="Salary",
    hue="Department",
    data=df)
```

3. Кількість людей посортованих по продуктивності

```python
sns.histplot(data=df, x='Productivity (%)')
```

4. Кількість людей (в процентовому розмірі) відсортованих по відділам

```python
plt.pie(df["Department"].value_counts(), labels=df["Department"].unique(), autopct="%1.1f%%")
plt.title("Department Distribution")
plt.show()
```

5. Коробкова діаграма для продуктивності співробітників за відділами

```python
plt.boxplot([df[df["Department"] == "Marketing"]["Productivity (%)"], df[df["Department"] == "IT"]["Productivity (%)"], df[df["Department"] == "Sales"]["Productivity (%)"]])

plt.xticks([1, 2, 3], ["Marketing", "IT", "Sales"])
plt.ylabel("Productivity (%)")

plt.show()
```

6. Розподіл співробітників за віком і зарплатою

```python
plt.scatter(df["Age"], df["Salary"])

plt.title("Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

7. Кількість зроблених проектів: чоловіки та жінки

```python
plt.bar(["Women", "Men"], [df[df["Gender"] == "Female"]["Projects Completed"].sum(), df[df["Gender"] == "Male"]["Projects Completed"].sum()])

plt.xlabel("Gender")
plt.ylabel("Number of Projects")

plt.show()
```

8. Скрипковий графік по відділах та зарплатні працівників

```python
sns.violinplot(x="Department", y="Salary", data=df)

plt.title("Salary Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```
