```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nodes_df = pd.read_csv("dining-table-partners-nodes.csv")
edges_df = pd.read_csv("dining-table-partners-edges.csv")
```

```python
G = nx.DiGraph()
```

```python
# Add nodes
for _, row in nodes_df.iterrows():
    G.add_node(row['Id'], label=row['Label'])

# Add edges
for _, row in edges_df.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=row['Weight'])
```

```python
# Create positions for node labels
pos = nx.spring_layout(G, k=0.3, iterations=20)  # Increase the scale factor

# Draw the graph
nx.draw(G, pos, with_labels=True, node_size=900, node_color='tab:blue', font_size=10, font_color='black')

# Draw node labels
node_labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8)

# Show the plot
plt.show()
```
