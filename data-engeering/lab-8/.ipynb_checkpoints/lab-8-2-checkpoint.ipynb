```python
import pandas as pd
import networkx as nx

import matplotlib.pyplot as plt
from matplotlib.colorbar import ColorbarBase
from matplotlib.colors import Normalize
```

```python
nodes_df = pd.read_csv("dining-table-partners-nodes.csv")
edges_df = pd.read_csv("dining-table-partners-edges.csv")
```

```python
G = nx.DiGraph()
```

```python
for i in range(len(nodes_df)):
    G.add_node(nodes_df["Id"][i], label=nodes_df["Label"][i])

for i in range(len(edges_df)):
    G.add_edge(edges_df["Source"][i], edges_df["Target"][i], id=edges_df["Id"][i], weight=edges_df["Weight"][i])
```

```python
print(G)
```

```python
print("Nodes:", G.nodes(data=True))
print("\n")
print("Edges:", G.edges(data=True))
```

```python
node_degrees = dict(G.degree())

# Create a color map based on node degrees
color_map = [node_degrees[node] for node in G.nodes]

# Normalize the color map to be in the range [0, 1]
norm = plt.Normalize(min(color_map), max(color_map))
node_colors = plt.cm.viridis(norm(color_map))

# Graph
fig, ax = plt.subplots(figsize=(8, 6))
pos = nx.circular_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=300, node_color=node_colors)
labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_labels(G, pos, labels=labels, font_size=6)

weights = nx.get_edge_attributes(G, 'weight').values()
options = {
    'edge_color': 'r',
    'width': [0.4 * i for i in weights],
}
nx.draw_networkx_edges(G, pos, arrows=True, **options)

# Colorbar
cbar = ColorbarBase(ax.inset_axes([1.0, 0.1, 0.05, 0.8]), cmap=plt.cm.viridis, norm=norm, orientation='vertical')
cbar.set_label('Шкала охочих гостей сидіти з людиною')

plt.show()
```
