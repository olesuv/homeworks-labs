```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df = pd.read_json("data.json")
df
```

```python
len(df)
```

```python
emails = df['email'].tolist()
emails
```

```python
mlb = MultiLabelBinarizer()
firm_matrix = mlb.fit_transform(df['profile'].apply(lambda x: [x['company']] if 'company' in x else []))

firm_df = pd.DataFrame(firm_matrix, columns=mlb.classes_)

firm_df.insert(0, 'Employee', df['username'])

firm_df
```

```python
mlb = MultiLabelBinarizer()
role_matrix = mlb.fit_transform(df['roles'])

role_df = pd.DataFrame(firm_matrix, columns=mlb.classes_)

role_df.insert(0, 'Employee', df['username'])

role_df
```

```python
df['dob'] = pd.to_datetime(df['profile'].apply(lambda x: x['dob']))

youngest_employee = df.loc[df['dob'].idxmin()]
oldest_employee = df.loc[df['dob'].idxmax()]

print("Наймолодший співробітник:")
print(youngest_employee[['profile', 'dob']])

print("\nНайстарший співробітник:")
print(oldest_employee[['profile', 'dob']])
```
