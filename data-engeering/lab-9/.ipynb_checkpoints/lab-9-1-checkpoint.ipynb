```python
import numpy as np
import librosa
```

```python
filename = "е4shot3500.wav"
y, sr = librosa.load(filename, sr=None)
```

```python
print(f'Частота дискретизації: {sr} Гц')
print(f'Кількість каналів: {y.shape[0]}')
print(f'Бітрейт: {librosa.get_samplerate(filename) * 8} біт/сек')
```

```python
num_samples = len(y)
duration = librosa.get_duration(y=y, sr=sr)

print(f'Кількість числових даних: {num_samples} зразків')
print(f'Тривалість звукового фрагменту: {duration:.2f} секунд')
```

```python
import matplotlib.pyplot as plt
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)

plt.figure(figsize=(12, 8))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Спектрограма звукового фрагменту')
plt.show()
```
