```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import numpy as np

import librosa
```

```python
filename = "е4shot3500.wav"
y, sr = librosa.load(filename)
```

```python
# Визначення кількості числових даних
num_samples = len(y)

# Обчислення тривалості звукового фрагменту в секундах
duration = librosa.get_duration(y=y, sr=sr)

print(f"Кількість числових даних: {num_samples}")
print(f"Тривалість звукового фрагменту: {duration} секунд")
```

```python
# Обчислення спектрограми
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# Побудова графіку спектрограми
plt.figure(figsize=(10, 4))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Спектрограма')
plt.show()
```

```python
# Вибір першої половини значень
half_length = len(y) // 2
first_half = y[:half_length]

# Збільшення амплітуди
amplitude_factor = 2.0
increased_amplitude = first_half * amplitude_factor

# Перевірка діапазону значень
if np.max(np.abs(increased_amplitude)) > 1.0:
    print("Увага: Діапазон значень вийшов за межі дозволених.")

# Побудова осцилограми для оригінального та збільшеного сигналів
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Осцилограма оригінального сигналу')

plt.subplot(3, 1, 2)
librosa.display.waveshow(increased_amplitude, sr=sr)
plt.title(f'Осцилограма збільшеного сигналу (в {amplitude_factor} рази)')

# Побудова спектрів
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
D_original = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
librosa.display.specshow(D_original, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Спектр оригінального сигналу')

plt.subplot(2, 1, 2)
D_increased = librosa.amplitude_to_db(np.abs(librosa.stft(increased_amplitude)), ref=np.max)
librosa.display.specshow(D_increased, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title(f'Спектр збільшеного сигналу (в {amplitude_factor} рази)')

plt.show()
```
