```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('Wine.csv')

X = df.drop(['Customer_Segment'], axis=1)
Y = df['Customer_Segment']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
scaler = StandardScaler()

x_scaled_train = scaler.fit_transform(X_train)
x_scaled_test = scaler.transform(X_test)
```

```python
for n_components in [3, 8, 13]:
    pca = PCA(n_components=n_components)
    x_pca_train = pca.fit_transform(x_scaled_train)
    x_pca_test = pca.transform(x_scaled_test)

    # Класифікатор DecisionTree
    clf_dt = DecisionTreeClassifier(max_depth=2, random_state=0)
    clf_dt.fit(x_pca_train, y_train)
    y_pred_dt = clf_dt.predict(x_pca_test)

    print(f'\nDecisionTree with {n_components} components:')
    print(f'Accuracy: {accuracy_score(y_test, y_pred_dt):.2f}')
    print(f'Confusion matrix:\n{confusion_matrix(y_test, y_pred_dt)}')
    
    # Класифікатор RandomForest
    clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
    clf_rf.fit(x_pca_train, y_train)
    y_pred_rf = clf_rf.predict(x_pca_test)

    print(f'\nRandomForest with {n_components} components:')
    print(f'Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}')
    print(f'Confusion matrix:\n{confusion_matrix(y_test, y_pred_rf)}')
```
