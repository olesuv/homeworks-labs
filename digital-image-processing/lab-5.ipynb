```python
from matplotlib import pyplot as plt
import numpy as np 
import random 
import math 
import cv2
```

```python
logo = cv2.imread('assets/square.png')
main = cv2.imread('assets/baloons.jpg')
```

```python
(logo_h, logo_w) = logo.shape[:2] 
(main_h, main_w) = main.shape[:2]

print(f"Size of logo image: {logo_h, logo_w}, size of main image: {main_h, main_w}") 
```

```python
# Визначення області призначення на основному зображенні
destination = main[main_h - logo_h - 10:main_h - 10, main_w - logo_w - 10:main_w - 10]

# Застосування водяного знаку до основного зображення
result = cv2.addWeighted(destination, 1, logo, 0.7, 0)

# Заміна області призначення на основному зображенні з модифікованою частиною
main[main_h - logo_h - 10:main_h - 10, main_w - logo_w - 10:main_w - 10] = result
```

```python
plt.imshow(main)
cv2.imwrite("assets/watermarked.jpg", main)
```

```python
RGB_img = cv2.cvtColor(main, cv2.COLOR_BGR2RGB) 
plt.imshow(RGB_img) 
plt.show()
```
