```python
from matplotlib import pyplot as plt
import cv2
```

```python
logo = cv2.imread('assets/square.png', -1)
logo = cv2.resize(logo, (0,0), fx=0.5, fy=0.5)

main = cv2.imread('assets/baloons.jpg')
```

```python
(logo_h, logo_w) = logo.shape[:2] 
(main_h, main_w) = main.shape[:2]

print(f"Size of logo image: {logo_h, logo_w}, size of main image: {main_h, main_w}") 
```

```python
x_offset = main_w - logo_w - 15
y_offset = main_h - logo_h - 185

y1, y2 = y_offset, y_offset + logo.shape[0]
x1, x2 = x_offset, x_offset + logo.shape[1]

alpha_logo = logo[:, :, 3] / 255.0
alpha_main = 1.0 - alpha_logo

for c in range(0, 3):
    main[y1:y2, x1:x2, c] = (alpha_logo * logo[:, :, c] + alpha_main * main[y1:y2, x1:x2, c])
```

```python
plt.imshow(cv2.cvtColor(main, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
