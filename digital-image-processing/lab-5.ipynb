```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
```

```python
main = cv.imread('assets/baloons.jpg', cv.IMREAD_GRAYSCALE)

kernel = np.array([[-1, -2], [2, 1]], dtype=np.float32)
kernel *= 5
```

```python
result = cv.filter2D(main, -1, kernel)
result
```

```python
plt.subplot(121), plt.imshow(cv.cvtColor(main, cv.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv.cvtColor(result, cv.COLOR_BGR2RGB)), plt.title('Filtered x5')
plt.show()
```
