```python
import matplotlib.pyplot as plt 
import numpy as np 
import cv2 as cv
```

```python
def read_image(file_path): 
    image = cv.imread(file_path) 
    return cv.cvtColor(image, cv.COLOR_BGR2RGB) 
 
def show_image(image, title='Image', cmap=None): 
    plt.figure(figsize=(6, 6)) 
    plt.imshow(image, cmap=cmap) 
    plt.title(title) 
    plt.axis('off') 
    plt.show() 
 
# Виділення RGB каналів 
def split_rgb_channels(image): 
    red_channel = image.copy() 
    red_channel[:, :, 1] = 0  # Видалити зелений канал 
    red_channel[:, :, 2] = 0  # Видалити синій канал 
    green_channel = image.copy() 
    green_channel[:, :, 0] = 0  # Видалити червоний канал 
    green_channel[:, :, 2] = 0  # Видалити синій канал 
    blue_channel = image.copy() 
    blue_channel[:, :, 0] = 0  # Видалити червоний канал 
    blue_channel[:, :, 1] = 0  # Видалити зелений канал 
    return red_channel, green_channel, blue_channel 
 
# Конвертація з RGB до HSV 
def convert_rgb_to_hsv(rgb_image): 
    return cv.cvtColor(rgb_image, cv.COLOR_RGB2HSV) 
 
# Конвертація з RGB до LAB 
def convert_rgb_to_lab(rgb_image): 
    return cv.cvtColor(rgb_image, cv.COLOR_RGB2LAB) 
```

```python
original_image = read_image('assets/baloons.jpg') 
show_image(original_image, 'Original Image') 
```

```python
# Відобразити RGB канали 
red_channel, green_channel, blue_channel = split_rgb_channels(original_image) 
show_image(red_channel, 'Red Channel') 
show_image(green_channel, 'Green Channel') 
show_image(blue_channel, 'Blue Channel')
```

```python
# Перетворення зображення 
hsv_image = convert_rgb_to_hsv(original_image) 
lab_image = convert_rgb_to_lab(original_image)

# Відображення оригінального зображення та перетворених зображень 
show_image(hsv_image, 'Converted to HSV', cmap='hsv') 
show_image(lab_image, 'Converted to LAB') 
```
