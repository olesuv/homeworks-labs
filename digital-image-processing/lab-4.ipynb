```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
from skimage import io
from scipy import ndimage
```

```python
path = './assets/girl.png'
img = io.imread(path, pilmode='L')

plt.imshow(img, cmap="gray")
```

```python
hist, bin_edges = np.histogram(img, bins='auto')
bin_centers = 10 * (bin_edges[:-1] + bin_edges[1:])
```

```python
im_min = img.min()
im_max = img.max()
im_mean = img.mean()

im_min, im_max, im_mean
```

```python
plt.plot(bin_centers, hist, lw=2)
```

```python
c = cv2.bitwise_not(img)
```

```python
d = 255 / np.log(1 + np.max(img)) 
log_image = d * (np.log(img + 1))
log_image = np.array(log_image, dtype=np.uint8)

hist, bin_edges = np.histogram(log_image, bins='auto')
bin_centers = 10 * (bin_edges[:-1] + bin_edges[1:])
```

```python
f, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(log_image, cmap='gray')
axes[1].plot(bin_centers, hist, lw=2)
```

```python
e = np.power(img, 0.8)
e = np.array(e, dtype=np.uint8)
```

```python
eps_image = log_image - 1 * (img + e) ** 0.3
eps_image = np.array(eps_image, dtype=np.uint8)

hist, bin_edges = np.histogram(eps_image, bins='auto')
bin_centers = 10 * (bin_edges[:-1] + bin_edges[1:])

f, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(eps_image, cmap='gray')
axes[1].plot(bin_centers, hist, lw=2)
```
