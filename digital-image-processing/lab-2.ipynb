```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
path = './assets/nasa.png'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
c = cv2.bitwise_not(img)
plt.imshow(c, cmap='gray')
```

```python
d = 255 / np.log(1 + np.max(img)) 
log_image = d * (np.log(img + 1))
log_image = np.array(log_image, dtype=np.uint8)

plt.imshow(log_image, cmap='gray')
```

```python
e = np.power(img, 0.8)
e = np.array(e, dtype=np.uint8)

plt.imshow(e, cmap='gray')
```

```python
eps_image = log_image - 1 * (img + e) ** 0.3
eps_image = np.array(eps_image, dtype=np.uint8)

plt.imshow(eps_image, cmap='gray')
```

```python
eps_image = log_image - 1 * (img + e) ** 0.4
eps_image = np.array(eps_image, dtype=np.uint8)

plt.imshow(eps_image, cmap='gray')
```

```python
eps_image = log_image - 1 * (img + e) ** 0.5
eps_image = np.array(eps_image, dtype=np.uint8)

plt.imshow(eps_image, cmap='gray')
```
