```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (256, 256))

plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title('Оригінальне зображення')
plt.axis('off')
plt.show()
```

```python
# 1. Згладжування зображення для зменшення шуму
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)
plt.figure(figsize=(6, 6))
plt.imshow(blurred_image, cmap='gray')
plt.title('Згладжене зображення (Gaussian Blur)')
plt.axis('off')
plt.show()
```

```python
# 2. Виділення контурів методом Кенні
low_threshold = 50
high_threshold = 150
edges = cv2.Canny(blurred_image, low_threshold, high_threshold)

plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap='gray')
plt.title('Контури (метод Кенні)')
plt.axis('off')
plt.show()
```

```python
# 3. Аналіз контурів та їх відображення
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)  # Конвертація до кольорового для відображення контурів
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)  # Зелені контури з товщиною 2

plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title('Зображення з виділеними контурами')
plt.axis('off')
plt.show()
```
