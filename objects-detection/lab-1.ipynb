```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Завантажуємо зображення
image = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Функція для показу зображення
def show_images(original_img, filtered_img, filter_name):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))

    # Оригінальне зображення
    axs[0].imshow(original_img, cmap='gray')
    axs[0].set_title('Original Image')
    axs[0].axis('off')

    # Відфільтроване зображення
    axs[1].imshow(filtered_img, cmap='gray')
    axs[1].set_title(f'{filter_name} Filtered Image')
    axs[1].axis('off')

    # Показати графіки
    plt.tight_layout()
    plt.show()

# Фільтр розмиття (Gaussian Blur)
def apply_gaussian_blur(image):
    blurred_image = cv2.GaussianBlur(image, (31, 31), 0)
    for _ in range(50):
        blurred_image = cv2.GaussianBlur(blurred_image, (31, 31), 0)
    return blurred_image

# Фільтр покращення чіткості зображення (Sharpen)
def apply_sharpen(image):
    kernel = np.array([[0, -1, 0],
                       [-1, 9, -1],
                       [0, -1, 0]])
    sharpened_image = cv2.filter2D(image, -1, kernel)
    return sharpened_image

# Медіанний фільтр
def apply_median_filter(image):
    median_image = cv2.medianBlur(image, 31)
    for _ in range(50):
        median_image = cv2.medianBlur(image, 31)
    return median_image

# Фільтр Ґабора
def apply_gabor_filter(image):
    kernel = cv2.getGaborKernel((21, 21), 8.0, np.pi / 4, 10.0, 0.5, 0, ktype=cv2.CV_32F)
    gabor_image = cv2.filter2D(image, -1, kernel)
    return gabor_image
```

```python
blurred_image = apply_gaussian_blur(image)
sharpened_image = apply_sharpen(image)
median_image = apply_median_filter(image)
gabor_image = apply_gabor_filter(image)
```

```python
show_images(image, image,'Original Image')
```

```python
show_images(image, blurred_image, 'Gaussian Blur')
```

```python
show_images(image, sharpened_image, 'Sharpen')
```

```python
show_images(image, median_image, 'Median Blur')
```
